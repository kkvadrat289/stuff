```python
def AND(gate, x, y):
    print("GATE %d AND %d %d" % (gate, x, y))

def OR(gate, x, y):
    print("GATE %d OR %d %d" % (gate, x, y))
    
def NOT(gate, x):
    print("GATE %d NOT %d" % (gate, x))
    
def OUT(gate, x):
    print("OUTPUT %d %d" % (gate, x))
```

```python
n = int(input())
gate = 3 * n
# 1) a AND b, range: 3n->4n
for vertex in range(0, n):
    AND(gate, vertex, vertex + n) 
    gate += 1
# 2) a OR b, range: 4n->5n
for vertex in range(0, n):
    OR(gate, vertex, vertex + n)
    gate += 1
# 3) NOT (a OR b), range: 5n->6n
for vertex in range(0, n):
    NOT(gate, 4 * n + vertex)
    gate += 1
# 4) (a AND b) OR NOT(a OR b) == NOT (a XOR b), range: 6n->7n
for vertex in range(0, n):
    OR(gate, 3 * n + vertex, 5 * n + vertex)
    gate += 1
# 5) NOT c, range: 7n->8n
for vertex in range(0, n):
    NOT(gate, 2 * n + vertex)
    gate += 1
# 6) NOT(a XOR b) OR NOT c, range: 8n->9n
for vertex in range(0, n):
    OR(gate, 6 * n + vertex, 7 * n + vertex)
    gate += 1
# 7) a XOR b, range: 9n->10n
for vertex in range(0, n):
    NOT(gate, 6 * n + vertex)
    gate += 1
# 8) (a XOR b) OR c, range: 10n->11n
for vertex in range(0, n):
    OR(gate, 9 * n + vertex, 2 * n + vertex)
    gate += 1
# 9) (NOT(a XOR b) OR NOT c) AND ((a XOR b) OR c), range: 11n->12n
for vertex in range(0, n):
    AND(gate, 8 * n + vertex, 10 * n + vertex)
    gate += 1
# 10) (a XOR b) AND c, range: 12n->13n
for vertex in range(0, n):
    AND(gate, 9 * n + vertex, 2 * n + vertex)
    gate += 1
# 11) (a AND b) OR ((a XOR b) AND c), range: 13n->14n
for vertex in range(0, n):
    OR(gate, 3 * n + vertex, 12 * n + vertex)
    gate += 1

#12) let's make zero, range: 14n
AND(gate, 2 * n, 7 * n)
    
out = 0
# 12) output x
for vertex in range(0, n):
    OUT(out, 11 * n + vertex)
    out += 1
for i in range(2):
    OUT(out, 14 * n)
    out += 1

# 13) output y
for vertex in range(0, n):
    OUT(out, 13 * n + vertex)
    out+=1
```
